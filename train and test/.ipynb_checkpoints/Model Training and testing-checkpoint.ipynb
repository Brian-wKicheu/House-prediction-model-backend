### Step 1: Import the libraries

```python
#pandas for data manipulation
#numpy math computing and arrays
#matplotlib pyplot for visualization
#seaborn visualization
# scikit-learn machine learning
#pickle for serialization
# we need standartisation because our data is of different scales i.e ones, tens, hundreds, thousands..... for same scale to penolize
# we RandomForestRegressor because data being used is continous(quantitative)
#metrics for our data to be on the same scale with features, the smaller the metrics the better perfomance of the metrics
```

```python
print("="*60)
print("Step 1. IMPORT DEPENDENCIES OR lIBRARIES")
print("="*60)

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
import pickle
```

### Step 2.1: Load the dataset

```python
# Step 2.1: Load the dataset
print("="*60)
print("Step 2 LOAD THE DATASET")
print("="*60)

df = pd.read_csv('house_data.csv')
print(f'\nDataset Loaded Successfully')
print(f'Total records: {len(df)}')
print(f'Total features: {len(df.columns)}')

```

### Step 2.2 Load overview

```python
#Step 2.2: Load overview
print("="*60)
print("LOAD OVERVIEW")
print("="*60)

print(f'\nFirst 5 row: ')
print(df.head())

print(f'\nDataset Info: ')
print(df.info())

print(f'\nStatistical summary: ')
print(df.describe())
```

### 2.3 Check Missing data

```python
# Check missing data
print("\n"+ "="*60)
print("Step 2.3 DATA QUALITY CHECK")
print("="*60)

print(f'\nMissing values per column')
print(df.isnull().sum())
```

### 2.4 Check for duplicates

```python
#check for duplicates
duplicates = df.duplicated().sum()
print(f'\nDuplicate rows : {duplicates}')
```

### Step 3.1: Exploratory Data Analysis

```python
#Step 3: Exploratory Data Analysis
print("\n"+ "="*60)
print("Step 3 EXPLORATIVE DATA ANALYSIS")
print("="*60)

# Distribution of the target variable (price)
print(f'\nPrice Statistics :')
print(f"Mean price: ${df['price'].mean():,.2f}")
print(f"Median price: ${df['price'].median():,.2f}")
print(f"Min price: ${df['price'].min():,.2f}")
print(f"Max price: ${df['price'].max():,.2f}")
print(f"Standard deviation price: $ {df['price'].std():,.2f}")
```

### 3.2 Check for correlation, factors contributing to rise or falling of the price

```python
#Correlation
print("\n"+ "="*60)
print("Correlation with price rising or falling")
print("="*60)

correlations = df.corr()['price'].sort_values(ascending=False)
print(f'\nFeatures correlation with price:')
print(correlations)

#Identify highly correlated features
high_corr = correlations[abs(correlations) > 0.5].drop('price')
print(f'\nHighly correlated features(| Correlations | > 0.5)')
```

### Step 4.1: Data Preprocessing

```python
# preprocessing
print("\n"+ "="*60)
print("Step 4 DATA PREPROCESSING")
print("="*60)

#Separate features and the target which is the price since it's the one we are predicting(dependent variable)
X = df.drop('price', axis = 1)
y = df['price'] #dependent variable

print(f'\nFeatures (x) shape {X.shape}:')
print(f'\nTarget (y) shape {y.shape}:')
```

### 4.2 Split data into training and testing sets

```python
# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 43)

print(f'\nTrain set size {len(X_train)} samples ({len(X_train)/len(X) * 100:.1f}%):')
print(f'\nTest set size {len(X_test)} samples ({len(X_test)/len(X) * 100:.1f}%):')
```

### 4.3: Feature Scaling (Standardisation)

```python
#Standardisation making values range from same scale 
#machine learning algorithms perform better when features have similar scale 
print("\n"+ "="*60)
print("FEATURE SCALING")
print("="*60)

scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.fit_transform(X_test)

print(f'Feature scaled using StandardScaler')
print(f'Formula: z = (x - mean)/std_deviation')

print(f"\nSample before scaling: \n\n{X_train.iloc[0].values}")
print(f"\nSample after scaling: \n\n{X_train_scaled[0]}")
```

### Step 5: Model Training


```python
print("\n"+ "="*60)
print("MODEL TRAINING")
print("="*60)

models = {}
#MODEL 1: linear Regression
print("\n[1/3] Training Linear Regression...")
lr_model = LinearRegression() #first algorithm we are going to use
lr_model.fit(X_train_scaled, y_train)
models['Linear Regression'] = lr_model
print("Linear Regression Trained")

#MODEL 2: RandomForestRegressor
print("\n[2/3] Training Random Forest...")
rf_model = RandomForestRegressor(n_estimators = 100, max_depth = 20, random_state =42, n_jobs = -1) #second algorithm 
rf_model.fit(X_train_scaled, y_train)
models['Random Forest Regressor'] = rf_model
print("Random Forest Trained")

#MODEL 3: GradientBoostingRegressor
print("\n[3/3] Training Gradient Boosting...")
gb_model = GradientBoostingRegressor(n_estimators = 100, learning_rate = 0.1, max_depth = 5, random_state =42) #third algorithm 
gb_model.fit(X_train_scaled, y_train)
models['Gradient Boosting Regressor'] = gb_model
print("GradientBoosting Training")

print("\nAll models have trained successfully")
```

### Step 6 Model Evaluation

```python
print("\n"+ "="*60)
print("STEP 6. MODEL EVALUATION")
print("="*60)

results = {}

for name, model in models.items():
    print("\n"+ "="*60)
    print(f"Evaluation: {name}")
    print("="*60)

    # prediction
    y_train_pred = model.predict(X_train_scaled)
    y_test_pred = model.predict(X_test_scaled)

    # Training metrics
    train_r2 = r2_score(y_train, y_train_pred)
    train_rmse = np.sqrt(mean_squared_error(y_train, y_train_pred))
    train_mae = mean_absolute_error(y_train, y_train_pred)

    # Testing metrics
    test_r2 = r2_score(y_test, y_test_pred)
    test_rmse = np.sqrt(mean_squared_error(y_test, y_test_pred))
    test_mae = mean_absolute_error(y_test, y_test_pred)

    #Results
    results[name] = {
        'test_r2': test_r2,
        'train_r2': train_r2,
        'test_rmse': test_rmse,
        'train_rmse': train_rmse,
        'test_mae': test_mae,
        'train_mae': train_mae,
    }

    print(f"\nTraining Perfomance:")
    print(f"\nR2 Score: {train_r2:.4f}")
    print(f"\nRMSE Score: {train_rmse:.2f}")
    print(f"\nMAE Score: {train_mae:.2f}")

    print(f"\nTesting Perfomance:")
    print(f"\nR2 Score: {test_r2:.4f}")
    print(f"\nRMSE Score: {test_rmse:.2f}")
    print(f"\nMAE Score: {test_mae:.2f}")

    overfit_diff = train_r2 - test_r2
    if overfit_diff > 0.1:
        print(f"Warning: Possible overfitting (difference: {overfit_diff:.4f})")
    else:
        print(f"Good generalization (difference: {overfit_diff:.4f})")

# 
```

### Step 7 Summary comparison

```python
print("\n"+ "="*60)
print("STEP 7. MODEL COMPARISON SUMMARRY")
print("="*60)
print(f"\n{'Model':<20}{'Test R2':<12}{'Test RMSE':<15}{'Test MAE':<15}")
print("-"*60)
for name, metrics in results.items():
    print(f"{name:<20}{metrics['test_r2']:<12,.4f}{metrics['test_rmse']:<14,.2f}{metrics['test_mae']:<15,.2f}")

#Now select best model
best_model_name = max(results, key=lambda x: results[x]['test_r2'])
best_model = models[best_model_name]
print(f"Best Model: {best_model_name}")
print(f"Test R2 Score: {results[best_model_name]['test_r2']:.4f}")
```

### Step 8. Saving Best Model through (Serialization)

```python
print("\n"+ "="*60)
print("STEP 8. SAVE BEST PERFOMING MODEL")
print("="*60)

#Save the best model
model_filename = 'house_price_model.pkl'
scaler_filename = 'scaler.pkl'

print(f"\nSaving the best model: {best_model_name}")
with open(model_filename, 'wb') as file: #basic file handling in write text file in binary
    pickle.dump(best_model, file)
print(f"Model saved as '{model_filename}'")

print(f"\nSaving scaler....")
with open(scaler_filename, 'wb') as file :
    pickle.dump(scaler, file)
print(f"Model saved as '{scaler_filename}'")

#save feature names
feature_names = X.columns.tolist()
with open('feature_names.pkl', 'wb') as file: 
    pickle.dump(feature_names, file)
print(f"Feature names saved as '{feature_names}'")

# Save model metadata
metadata = {
    'model_name': best_model_name,
    'test_rmse':results[best_model_name]['test_rmse'],
    'test_mae':results[best_model_name]['test_mae'],
    'features': feature_names,
    'training_samples': len(X_train),
    'test_samples': len(X_test)
}

with open('model_metadata.pkl', 'wb') as file:
    pickle.dump(metadata, file)
print(f"Metadata saved")

print("\n"+ "="*60)
print("MODEL TRAINING COMPLETE")
print("="*60)

print(f"\nFiles created")
print(f" 1. {model_filename} - Trained model")
print(f" 2. {scaler_filename} - Feature Scaler")
print(f" 3. feature_names.pkl - Feature names")
print(f" 4. model_metadata.pkl - Model Information")
```
